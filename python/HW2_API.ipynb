```python
from requests import get
from time import sleep

api_url = 'https://freetestapi.com/api/v1/weathers'

response = get(api_url)

if response.status_code == 200:
    data = response.json()  # แปลง response เป็น JSON

    character = []
    for response_js in data:
        city = response_js["city"]
        temperature = response_js["temperature"]
        weather_description = response_js["weather_description"]
        result = [city, temperature, weather_description]
        character.append(result)
        sleep(2)  #break 2 seconds

    print(character)

else:
     print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
header = ["City, Temperature, Weather_description"]

import csv

with open("weathers.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(character)
```
